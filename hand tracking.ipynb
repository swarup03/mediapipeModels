```python
import cv2
# ! pip install mediapipe
import mediapipe as mp
import time
```

```python
cap = cv2.VideoCapture(1)
```

```python
mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils

ptime =0
ctime = 0
```

```python
# this is a code for simply show the vido using camara
while True:
    success, img = cap.read()
    imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)
#     print(results.multi_hand_landmarks) # if i will tack my hand in front of the camara i will see the change in output
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            #now accessing the points of hands
            for id,lm in enumerate(handLms.landmark):
#                 print(id,lm) print hand points number and their x,y,z
                h,w,c = img.shape
                cx, cy =int(lm.x*w),int(lm.y*h)
                print(id,cx,cy)
                if id == 4:
                    cv2.circle(img,(cx,cy),15,(255,0,255),cv2.FILLED)
            mpDraw.draw_landmarks(img,handLms,mpHands.HAND_CONNECTIONS)
    
    ctime = time.time()
    fps = 1/(ctime-ptime)
    ptime = ctime
    
    cv2.putText(img,str(int(fps)),(10,70),cv2.FONT_HERSHEY_COMPLEX,3,(255,255,0),2)
    cv2.imshow("Image",img)
    cv2.waitKey(1)
```
